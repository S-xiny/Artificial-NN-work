```python
# -*- coding: utf-8 -*-
from fuzz.model.dbn import DBN
from fuzz.model.sae import SAE
from fuzz.model.cnn import CNN

def get_model(name):
    if name == 'dbn':
        # DBN
        para = {'struct': [784,400,100,10],
                'h_type': ['Gaussian', 'Binary'],
                'output_func': 'Affine',
                'v_type': ['Binary', 'Binary'],
                'dropout': 0.0,
                'task': 'cls',
                'reshape_size': [28,28],
                'flatten': True}
        return DBN(**para)

model = get_model('dbn')
model.load_mnist('../data', 128)
model.run(e = 10, pre_e = 3, load = '', n_sampling = 9)
```
